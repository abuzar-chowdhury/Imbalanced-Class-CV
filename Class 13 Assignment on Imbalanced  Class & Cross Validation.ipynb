# Over Sampling

```python
df1.head()
```

```python
df1.phishing_status.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler

over = RandomOverSampler(random_state=100)
```

```python
over_x, over_y = over.fit_resample(x,y)
```

```python
over_x.shape
```

```python
over_y.shape
```

```python
over_y.value_counts()
```

# Cross Validation

```python
# Hold Out CV
from sklearn.model_selection import train_test_split as tts

xtrain, xtest, ytrain, ytest = tts(over_x, over_y, train_size=.75, random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
xtest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(xtrain, ytrain)
```

```python
clf.score(xtest, ytest)
```

# KFold CV

```python
from sklearn.model_selection import KFold, cross_val_score, cross_validate

kfold = KFold(n_splits=5) # each fold contains 20% data
```

```python
result = cross_val_score(xgb, over_x, over_y, cv=kfold) # over_x & over_y from over sampling
result
```

```python
result.mean() # avg accuracy
```

```python
cross_validate = cross_validate(xgb, over_x, over_y, cv=kfold)
cross_validate
```
